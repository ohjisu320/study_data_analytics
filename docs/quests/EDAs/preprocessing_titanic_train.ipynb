## 결측치 및 이상치 처리

### 결측치 처리

```python
df_titanic.isnull().sum()
```

#### 1. 'Age' 열에서 결측치를 평균 나이로 채우기.

```python
mean = df_titanic['Age'].mean()
```

```python
df_titanic['Age'].fillna(mean, inplace=True)
```

```python
df_titanic.isnull().sum()
```

#### 2. 'Cabin' 열에 결측치가 있는 행을 찾고, 해당 행들의 평균 'Fare' 계산하기.

```python
df_titanic['Cabin_null'] = df_titanic['Cabin'].isnull()
```

```python
condition_cabin = 'Cabin_null==True'
df_titanic_Cabin_null = df_titanic.query(condition_cabin)
df_titanic_Cabin_null['Fare'].mean()
```

3. 'Embarked' 열에서 결측치가 있는 행을 찾고, 가장 많이 등장하는 항구로 채우기.

```python
df_titanic['Embarked_null'] = df_titanic['Embarked'].isnull()
```

```python
Embarked_top = df_titanic['Embarked'].describe().loc['top'] # S
```

```python
condition_embarked = 'Embarked_null==True'
df_titanic.query(condition_embarked)
df_titanic.fillna(Embarked_top, inplace=True)
df_titanic.loc[61:829]
```

### 이상치 처리

#### 4. 'Fare' 열에서 평균보다 3표준편차 이상 높은 이상치를 가진 행들을 찾기.

```python
over_outlier = df_titanic['Fare'].std()*3
over_outlier
```

```python
condition = f"Fare >= {over_outlier}"
df_titanic.query(condition)
```

#### 5. 'SibSp' (형제 자매/배우자 수)가 5 이상인 이상치를 가진 행들을 찾기.

```python
df_titanic['SibSp'].describe()
```

```python
condition = 'SibSp >= 5'
df_titanic.query(condition).sort_values(by='SibSp')
```

#### 6. 'Parch' (부모/자녀 수) 열에서 0이 아닌 값을 가진 행들의 평균 'Age' 계산하기.

```python
condition = 'Parch != 0'
df_parch = df_titanic.query(condition)
df_parch['Age'].mean()
```

#### 7. 'Fare'가 0인 행들을 찾고, 이를 평균 'Fare'로 채우기.

```python
def change_fare(x):
    if x==0 :
      x=df_titanic['Fare'].mean()
    return x
pass
df_titanic["Fare"] = df_titanic["Fare"].apply(change_fare)
df_titanic["Fare"].value_counts
```

#### 8. 'Pclass' (티켓 등급)가 1등급인 승객들 중 'Cabin' 열의 결측치 비율 계산하기.

```python
condition = 'Pclass == 1'
df_firstclass = df_titanic.query(condition)
round(df_firstclass['Cabin'].isnull().sum()*100/df_firstclass['Cabin'].describe().loc['count'], 2)
```
