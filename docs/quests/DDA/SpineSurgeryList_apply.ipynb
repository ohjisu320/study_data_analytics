```python
df_SSL = pd.read_csv('SpineSurgeryList.csv')
```

```python
def cal_bmi(params):
  result =params['체중']/(params['신장']*0.1)**2
  return round(result*100, 1)
df_SSL['bmi']=df_SSL[["신장", "체중"]].apply(cal_bmi,  axis=1)
pass
```

```python
def surgery_hours(params):
  result_hours =int(params['수술시간']//60)
  return pd.to_datetime(result_hours, format="%H")
df_SSL['수술시간_datetime_hours']=df_SSL['수술시간'].apply(surgery_hours)
pass
```

```python
def surgery_hours(params):
    try :
        result =int(params//60)
    except :
        result = 'nan'
    return result
df_SSL['수술시간_datetime_hours']=df_SSL['수술시간'].apply(surgery_hours)
```

```python
def surgery_minutes(params):
    try :
        result =int(params%60)
    except :
        result = 'nan'
    return result
df_SSL['수술시간_datetime_minutes']=df_SSL['수술시간'].apply(surgery_minutes)
```
